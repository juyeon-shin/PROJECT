### 1.4 수화인 주소 상위 지역 추출

```python
def k_count_top(fc_k):
  
  #수화인주소 카운트 그래프
  fc_k_c = fc_k.groupby(['수화인주소1,2'], as_index=False).count()
  fig = px.bar(fc_k_c, x='수화인주소1,2', y='창고코드', title='FC 곤지암 지역 물량 횟수')
  #fig.show()

  fc_k_c_s = fc_k_c.sort_values(by='창고코드', ascending=False)
  fc_k_c_s = fc_k_c_s.reset_index(drop=True)
  fig = px.bar(fc_k_c_s, x='수화인주소1,2', y='창고코드', title='FC 곤지암 지역 물량 횟수 정렬')
  fig.show()

  #상위 5 지역 추출
  fc_k_c_list = []
  fc_k_sc_list = []

  fc_k_r = []
  for i in range(10):
    fc_k_c_list.append(fc_k_c_s['수화인주소1,2'][i])
    fc_k_sc_list.append(fc_k_c_s['창고코드'][i])

  fc_k_r= pd.DataFrame()
  fc_k_r['지역'] = fc_k_c_list
  fc_k_r['물량'] = fc_k_sc_list

  return fc_k_r
```
