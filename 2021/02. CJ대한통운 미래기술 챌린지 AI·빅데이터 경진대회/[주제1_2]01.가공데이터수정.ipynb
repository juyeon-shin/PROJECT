# 1. 데이터 변수 추가 및 가공

```python
kyonggi=kyonggi[kyonggi.columns.difference(['배달터미널코드', '배달터미널소분류코드','권역구분'])]
gyeongnam=gyeongnam[gyeongnam.columns.difference(['배달터미널코드', '배달터미널소분류코드','권역구분'])]
```

```python
def function(df):
  df['주문시간']=df['주문시간'].astype('str')
  
def function1(df):
  for i in range(0,len(df['주문시간'])):
    df['주문시간'][i]=df['주문시간'][i].zfill(6)[0:2]



```

```python

def function2(df):
  for i in range(0,len(df['품목코드'])):
    df['품목코드'][i]=df['품목코드'][i][0:1]
```

```python

function(kyonggi)
function(gyeongnam)

function1(kyonggi)
function1(gyeongnam)

function2(kyonggi)
function2(gyeongnam)

```

```python
kyonggi['날짜']=pd.to_datetime(kyonggi['날짜'])
gyeongnam['날짜']=pd.to_datetime(gyeongnam['날짜'])

kyonggi=kyonggi[kyonggi['품목금액']!='N']
gyeongnam=gyeongnam[gyeongnam['품목금액']!='N']

kyonggi['품목금액']=kyonggi['품목금액'].astype('float')
gyeongnam['품목금액']=gyeongnam['품목금액'].astype('float')

kyonggi['주문시간']=kyonggi['주문시간'].astype('float')

gyeongnam['주문시간']=gyeongnam['주문시간'].astype('float')
```

```python
kyonggi.reset_index(inplace=True)
gyeongnam.reset_index(inplace=True)
```

```python
kyonggi['주문시간']=kyonggi['주문시간'].astype('int')
gyeongnam['주문시간']=gyeongnam['주문시간'].astype('int')
```

```python


#주문시간 수정하기
def function3(df):
    
    
    for i in range(len(df['주문시간'])):
      
          
          if df['주문시간'][i]<6:
            df['주문시간'][i]=str('새벽')
          elif 6<=df['주문시간'][i]<9:
            df['주문시간'][i]=str('아침')
          elif 9<=df['주문시간'][i]<12:
            df['주문시간'][i]=str('오전')
          elif 12<=df['주문시간'][i]<14:
            df['주문시간'][i]=str('점심')
          elif df['주문시간'][i]<17:
            df['주문시간'][i]=str('오후')
          elif df['주문시간'][i]<19:
            df['주문시간'][i]=str('저녁')
          else:
            df['주문시간'][i]=str('밤')
    

```

```python
function3(kyonggi)
function3(gyeongnam)
```

# 2. 데이터 변수 조건 생성 & count/mean한 새 변수 추출

```python
# 수화인주소포함

kyonggi1=kyonggi.groupby(['날짜','주문시간','수화인주소1,2','isHoliday','isAnniversary','isHoliday_yest','incDec','요일','품목코드'],as_index=False).count()

kyonggi1.rename(columns={'품목금액':'물량'},inplace=True)


kyonggi1['품목금액']=kyonggi.groupby(['날짜','주문시간','수화인주소1,2','isHoliday','isAnniversary','isHoliday_yest','incDec','요일','품목코드'],as_index=False).mean()['품목금액']
kyonggi1['강수량']=kyonggi.groupby(['날짜','주문시간','수화인주소1,2','isHoliday','isAnniversary','isHoliday_yest','incDec','요일','품목코드'],as_index=False).mean()['강수량']
kyonggi1['최고기온']=kyonggi.groupby(['날짜','주문시간','수화인주소1,2','isHoliday','isAnniversary','isHoliday_yest','incDec','요일','품목코드'],as_index=False).mean()['최고기온']
kyonggi1['최저기온']=kyonggi.groupby(['날짜','주문시간','수화인주소1,2','isHoliday','isAnniversary','isHoliday_yest','incDec','요일','품목코드'],as_index=False).mean()['최저기온']
kyonggi1['평균기온']=kyonggi.groupby(['날짜','주문시간','수화인주소1,2','isHoliday','isAnniversary','isHoliday_yest','incDec','요일','품목코드'],as_index=False).mean()['평균기온']

#품목코드를 기준으로 진행

#수화인주소포함

gyeongnam1=gyeongnam.groupby(['날짜','주문시간','수화인주소1,2','isHoliday','isAnniversary','isHoliday_yest','incDec','요일','품목코드'],as_index=False).count()

gyeongnam1.rename(columns={'품목금액':'물량'},inplace=True)


gyeongnam1['품목금액']=gyeongnam.groupby(['날짜','주문시간','수화인주소1,2','isHoliday','isAnniversary','isHoliday_yest','incDec','요일','품목코드'],as_index=False).mean()['품목금액']
gyeongnam1['강수량']=gyeongnam.groupby(['날짜','주문시간','수화인주소1,2','isHoliday','isAnniversary','isHoliday_yest','incDec','요일','품목코드'],as_index=False).mean()['강수량']
gyeongnam1['최고기온']=gyeongnam.groupby(['날짜','주문시간','수화인주소1,2','isHoliday','isAnniversary','isHoliday_yest','incDec','요일','품목코드'],as_index=False).mean()['최고기온']
gyeongnam1['최저기온']=gyeongnam.groupby(['날짜','주문시간','수화인주소1,2','isHoliday','isAnniversary','isHoliday_yest','incDec','요일','품목코드'],as_index=False).mean()['최저기온']
gyeongnam1['평균기온']=gyeongnam.groupby(['날짜','주문시간','수화인주소1,2','isHoliday','isAnniversary','isHoliday_yest','incDec','요일','품목코드'],as_index=False).mean()['평균기온']


#품목코드를 기준으로 진행


```
