# 기념일

```python

xmlUrl='http://apis.data.go.kr/B090041/openapi/service/SpcdeInfoService/getAnniversaryInfo'
service_key=unquote('b8sLGVKSMcrfCRS5JTtLN8iiuZrgcpTjGbMIU8UUnlE85Z5hif86%2FWb8tFU698WvQWmE01EMuSlXbo%2B92wxVpg%3D%3D'
)
```

```python
queryParams = '?' + urlencode({quote_plus('ServiceKey') : service_key, 
                               quote_plus('pageNo') : '1',
                               quote_plus('numOfRows') : '10', 
                               quote_plus('solYear') : '2021',
                               quote_plus('solMonth') : '06'
                              })
# quote_plus('solYear') : '2021',  quote_plus('solMonth') : '03'
```

```python
response = requests.get(xmlUrl + queryParams).text.encode('utf-8')
xmlobj = bs4.BeautifulSoup(response, 'lxml-xml')
print(xmlobj)
```

```python
pd.DataFrame(xmlobj)
```

```python
rows = xmlobj.findAll('item')
```

```python
columns=rows[0].find_all()
columns
```

```python
print(columns[0].name)
print(columns[0].text)
```

```python
rowList = []
nameList = []
columnList = []

rowsLen = len(rows)
for i in range(0, rowsLen):
    columns = rows[i].find_all()
    
    columnsLen = len(columns)
    for j in range(0, columnsLen):
        # 첫 번째 행 데이터 값 수집 시에만 컬럼 값을 저장한다. (어차피 rows[0], rows[1], ... 모두 컬럼헤더는 동일한 값을 가지기 때문에 매번 반복할 필요가 없다.)
        if i == 0:
            nameList.append(columns[j].name)
        # 컬럼값은 모든 행의 값을 저장해야한다.    
        eachColumn = columns[j].text
        columnList.append(eachColumn)
    rowList.append(columnList)
    columnList = []    # 다음 row의 값을 넣기 위해 비워준다. (매우 중요!!)
    
```

```python
result_03 = pd.DataFrame(rowList, columns=nameList)
result_03.head()
```

```python
result_04 = pd.DataFrame(rowList, columns=nameList)
result_04
```

```python
result_05 = pd.DataFrame(rowList, columns=nameList)
result_05
```

```python
result_06 = pd.DataFrame(rowList, columns=nameList)
result_06
```

```python
#merge
df1=pd.concat([result_03,result_04,result_05,result_06],axis=0)

df1.to_csv("C:/Users/ns451/Documents/프로젝트/대한통운 프로젝트[공모전]/holiday.csv",header=True,index=False)
```
