## 2.3 지역 데이터에 코로나, 휴일여부 결합

### 2.3.1 코로나,휴일 데이터 

```python
corona = pd.read_csv('/content/drive/Shareddrives/2021 CJ 대한통운 공모전/data/코로나 data/코로나_확진자_데이터.csv')
corona

def co_c(i):
  co_g = corona[corona['gubun'] == i]
  co_g = co_g.reset_index(drop=True)
  co_g['createDt'] = pd.to_datetime(co_g['createDt'], format='%Y-%m-%d', errors='raise')
  co_g['날짜'] = co_g['createDt']
  co_g['날짜'] = co_g['날짜'].astype('str')
  co_g['일일 코로나 확진자'] = co_g['incDec'] 
  co_g['주'] = co_g['createDt'].dt.week
  co_g = co_g.drop(['createDt','incDec'], axis=1)
  
  return co_g
```

```python
holiday = pd.read_csv('/content/drive/Shareddrives/2021 CJ 대한통운 공모전/data/휴일 data/final_휴일_1.csv')
holiday
#holiday['날짜'] = pd.to_datetime(holiday['날짜'], format='%Y-%m-%d', errors='raise')
```
