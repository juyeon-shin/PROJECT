```python


#data_sf = data[data['clac_hlv_nm'] != '스포츠패션']
data_sf = df[df['clac_hlv_nm'] != '스포츠패션']
data_sf.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le_1 =LabelEncoder()
le_2 =LabelEncoder()

from kmodes.kmodes import KModes
```

```python
#인코딩 적용
data_clac = data_sf[['rct_no','clac_mcls_nm']]
data_clac_en = data_clac.apply(le_1.fit_transform)
#data_clac_en = pd.concat([data_clac,data_sf[['buy_am','buy_ct']]],axis=1)
data_clac_en
```

```python
#k-modes 클러스터링
n_clusters = 10
km_cao = KModes(n_clusters=n_clusters, init = "Cao", n_init = 1, verbose=1)
fitClusters_cao = km_cao.fit_predict(data_clac_en)
```

```python
#클러스터링 중심점 
clusterCentroidsDf = pd.DataFrame(km_cao.cluster_centroids_)
clusterCentroidsDf.columns = data_clac_en.columns
clusterCentroidsDf
```

```python
data_clac_en['cluster'] = fitClusters_cao
data_clac_en
```

```python
en_df = pd.concat([data_clac,data_clac_en],axis=1)
en_df.columns = ['rct_no','clac_mcls_nm','rct_no_en','clac_mcls_nm_en','cluster']
#en_df.drop_duplicates(['pd_nm','pd_nm_en'])
en_df
```

```python
def cluster_plot(data,n):
    filtered = data[data['cluster']==n]
    #fil_hlv = filtered.groupby(['clac_hlv_nm']).count()
    #fil_mcls = filtered.groupby(['clac_mcls_nm']).count()
    #fil_pd = filtered.groupby(['clac_mcls_nm']).count()

     #plt.figure(figsize=(10,6))
    #sns.barplot(x = fil_hlv.index,y='X55',data=fil_hlv)
    #plt.title('cluster %d 군집 내 기업 수 : %d'%(i,len(fil_hlv)))
    return filtered
```

```python
def clusters_count(n_clusters):
    clu = cluster_plot(en_df,n_clusters)
    c_c = clu.groupby(['clac_mcls_nm']).count()
    print('clusters %d count: '%n_clusters,len(clu))
    return c_c
```
