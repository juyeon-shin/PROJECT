```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 사용을 위해서 세팅
from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/NGULIM.TTF"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)
```

```python
hd = pd.read_csv('지난 1년 동안 휴가 기간 동안 여가활동 (복수응답) 1028379_20220708003022.csv',encoding='cp949')
hd = hd.drop(['단위','2016 년','Unnamed: 8'],axis=1)
hd
```

```python
hd['통계분류'].unique()
```

```python
def top10(target):
    data = hd[hd['통계분류'] == target].head(10)
    index_list = data['항목']
    data.index = index_list
    data = data.drop(['통계분류','항목'],axis=1)

    for i in range(len(data)):
        sns.lineplot(x=data.columns,y=data.iloc[i,:], data=data.T)
        plt.xticks(rotation=45)
        plt.title('%s'%data.index[i])
        plt.show()
        
```

```python
top10('전체')
```
