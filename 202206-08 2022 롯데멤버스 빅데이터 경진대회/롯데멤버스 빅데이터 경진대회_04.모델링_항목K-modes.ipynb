```python
from kmodes.kmodes import KModes
```

```python
sport_rct_df_20 = sport_rct_df_20[sport_rct_df_20['clac_hlv_nm'] == '스포츠패션']
```

```python
all_2 = sport_rct_df_20[['clac_hlv_nm','clac_mcls_nm','pd_nm']]
all_2
```

```python
all_20 = all_2.apply(le.fit_transform)
```

```python
all_20
```

```python
#엘보우 포인트 찾기

cost = []
for num_clusters in list(range(1,20)):
    kmode = KModes(n_clusters=num_clusters, init = "Cao", n_init = 1, verbose=1)
    kmode.fit_predict(all_2)
    cost.append(kmode.cost_)
```

```python
y = np.array([i for i in range(1,20,1)])
plt.plot(y,cost)
```
