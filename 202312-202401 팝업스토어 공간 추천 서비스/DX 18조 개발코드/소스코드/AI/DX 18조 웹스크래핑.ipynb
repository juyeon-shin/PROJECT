```python
names = []
prices = []
addresses = []
units = []

while True:
    bh = crawler.execute_script('return document.body.scrollHeight') #브라우저 처음 높이 (스크롤 시작)
    time.sleep(1)
    crawler.execute_script('window.scrollTo(0, document.body.scrollHeight)')
    time.sleep(1)
    ah = crawler.execute_script('return document.body.scrollHeight')
    if ah == bh:
        break
    bh = ah
    
infos = crawler.find_elements(By.CSS_SELECTOR, ".SpaceItemstyle__Container-sc-1wddfv-5.kjjHma")
print(infos)
for info in infos:
    try:
        # 장소 이름 추출
        name = info.find_element(By.CSS_SELECTOR, "p.SpaceItemstyle__SpaceName-sc-1wddfv-13.jSNaYt").text
        names.append(name)

        # 주소 추출
        address = info.find_element(By.CSS_SELECTOR, "p.SpaceItemstyle__Address-sc-1wddfv-15.ijNkov").text
        addresses.append(address)

        # 가격 추출
        price = info.find_element(By.CSS_SELECTOR, "span.SpaceItemstyle__Price-sc-1wddfv-18.eXIEPQ").text
        prices.append(price)
        
        # 단위 추출
        unit = info.find_element(By.CSS_SELECTOR, "span.SpaceItemstyle__RuleType-sc-1wddfv-19.bblVi").text
        units.append(unit)
                
        
        print(names, addresses, prices, units)
        
    except Exception as e:
        print('Exception:', e)
```
