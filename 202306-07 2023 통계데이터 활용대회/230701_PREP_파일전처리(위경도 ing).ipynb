```python
import pandas as pd
import glob
```

```python
data_list = glob.glob('*.xlsx*')
print(data_list)
```

```python
for i in range(len(data_list)):
    data = pd.read_excel(data_list[i])
```

```python
data = pd.read_excel(data_list[2])
df = data[data['Latitude'].isnull() == True]
print(df.index[0])
df
```

```python
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')

# 위도, 경도 반환하는 함수
def geocoding(address):
    try:
        geo = geo_local.geocode(address)
        x_y = [geo.latitude, geo.longitude]
        return x_y

    except:
        return [0,0]
```

```python
latitude = []
longitude =[]

address = df['Full Address']

for i in address:
    latitude.append(geocoding(i)[0])
    longitude.append(geocoding(i)[1])
    
lat_log = pd.DataFrame([latitude,longitude]).T
lat_log.columns = ['Latitude','Longitude']
```

```python
lat_log['관광지명'] = df['관광지명']
lat_log
```

```python
lat_log.index = df['관광지명'].index
```
