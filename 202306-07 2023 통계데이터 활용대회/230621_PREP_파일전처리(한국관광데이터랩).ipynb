    - 외국인 관심 관광지 (영어, 일어, 중어)

```python
import zipfile
import os
import glob

import pandas as pd
import numpy as np
```

```python
list_zip = glob.glob('data/**.zip')
print(list_zip)

for i in list_zip:
    zipfile.ZipFile(i).extractall()
```

```python
list_csv = glob.glob('**.csv')
print(list_csv)
```

```python
def mk_sido(data):
    add_list = []
    for i in range(len(data)):
        add_list.append(data['주소'][i][:2])
    data['시도'] = add_list
    return data

def mk_sort(data):
    data = data.sort_values(data.columns[4], ascending=False, ignore_index=True)
    return data

def mk_df(name, sido = True):
    df = pd.DataFrame()
    list_csv = glob.glob(name)
    for i in list_csv:
        data = pd.read_csv(i, encoding='cp949')
        df = pd.concat([df,data],axis=0, ignore_index=True)
        
    
    if sido == True:
        df = mk_sido(df)
        df = mk_sort(df)
        df.columns = ['순위','관광지명','주소','분류','검색수','시도']
       
    return df
```

```python
pop_out = mk_df('*_인기관광지_외지인.csv')
pop_in = mk_df('*_인기관광지_현지인.csv')
pop_all = mk_df('*_인기관광지_전체.csv')

pop_local = mk_df('*_내국인 *.csv',sido=False)
pop_for = mk_df('*_외국인 *.csv',sido=False)
```

```python
pop_out['종류'] = '외지인'
pop_in['종류'] = '현지인'
pop_all['종류'] = '전체'

pop = pd.DataFrame()
pop = pd.concat([pop,pop_out,pop_in,pop_all],axis=0)
pop
```
