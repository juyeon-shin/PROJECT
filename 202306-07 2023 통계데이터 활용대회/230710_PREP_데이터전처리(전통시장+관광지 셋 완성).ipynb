```python
import pandas as pd
import glob
```

```python
data_list = glob.glob('market_*.csv')
print(data_list)
```

```python
final_market = pd.read_csv('final_market.csv')

market_pop = pd.read_csv('market_pop.csv')
# market_pop_lofor = pd.read_csv('market_pop_lofor.csv')

# market_pop_10 = pd.read_csv('market_pop_10.csv')
# market_pop_lofor_10 = pd.read_csv('market_pop_lofor_10.csv')
```

```python
print(len(final_market))
final_market.head(3)
```

```python
def make_merge(data):
    df = pd.DataFrame(data['시장(상점가명)'])
    
    for i in range(len(data_list)):
        origin = pd.read_csv(data_list[i])
        mer = pd.DataFrame(origin.groupby('시장(상점가명)').count()['Latitude'])
        mer.columns = data_list[i:i+1]
        
        df = pd.merge(left=df, right=mer,how='left', on='시장(상점가명)')
        
    return df
```

```python
merge_data = make_merge(final_market)
print(merge_data.isnull().sum())
merge_data = merge_data.fillna(0)
print(merge_data.isnull().sum())
merge_data
```

```python
merge_data.columns = [ list(merge_data.columns)[i].split('.')[0] for i in range(len(merge_data.columns))]
merge_data
```

```python
final_market_pop = pd.merge(left=final_market, right = merge_data, how='left', on='시장(상점가명)')
final_market_pop
```

```python
final_market_pop.to_csv('final_market_pop.csv',index=False)
```
