```python
df = pd.read_csv('소상공인시장진흥공단_전통시장현황_20220603.csv',encoding='cp949')
df.head(5)
```

```python
sido = []
for i in range(len(df)):
    sp = df['주소(도로명)'][i].split(' ')[0]
    sido.append(sp)
    
df['시도']=sido
df.head(5)
```

```python
df_cap = pd.concat([df[df['시도'] == '서울특별시'],df[df['시도'] == '경기도'],df[df['시도'] == '인천광역시']],axis=0)
df_cap.head(5)
```

```python
df_cap['시도'].unique()
```

```python
df_cap.to_csv('capital_market.csv', index=False)
```
