```python
import folium

def map_reset(lat,lon):
    new_map = folium.Map([lat,lon],zoom_start=10, tiles='cartodbpositron')
    return new_map

def map_marker(new_map,df,color,radius):
    df = df[df['Latitude'].notnull() == True]
    for i in range(len(df)):
        try:
            folium.Circle([df['Latitude'][i],df['Longitude'][i]],
                      color='#fffggg',radius=radius,
                     fill_color= color).add_to(new_map)
        except:
            pass
    return new_map
```

```python
new_map = map_reset(capital_market['Latitude'].mean(),capital_market['Longitude'].mean())
map_marker(new_map,capital_market,'#fffggg',radius=1000)
```
