```python
import pandas as pd
import numpy as np

import glob
```

```python
data_list = glob.glob('*.xlsx')
print(data_list)

data_inout = ['pop_all.xlsx','pop_in.xlsx','pop_out.xlsx']
data_lofor = ['pop_for.xlsx', 'pop_local.xlsx']
```

```python
pd.read_excel(data_inout[0])
```

```python
import folium

def map_reset(lat,lon):
    new_map = folium.Map([lat,lon],zoom_start=10, tiles='cartodbpositron')
    return new_map

def map_marker(new_map,df,color,radius):
    df = df[df['Latitude'].notnull() == True]
    for i in range(len(df)):
        try:
            folium.Circle([df['Latitude'][i],df['Longitude'][i]],
                      color='#fffggg',radius=radius,
                     fill_color= color).add_to(new_map)
        except:
            pass
    return new_map
```

```python
new_map = map_reset(37.477092,126.865575)
new_map

color = ['green','red','blue'] #전체, 현지인, 외지인

for i,v in enumerate(data_inout):
    df = pd.read_excel(data_inout[i])#,encoding='cp949')
    new_map = map_marker(new_map,df,color[i],radius=500)
new_map
```

```python
print(data_lofor)
new_ = map_reset(37.477092,126.865575)
new_

color = ['blue','red',] #외국인, 내국인

for i,v in enumerate(data_lofor):
    df = pd.read_excel(data_lofor[i])#,encoding='cp949')
    new_ = map_marker(new_,df,color[i],radius=500)
new_
```
