```python
market = pd.read_csv('market.csv')
market.head(3)
```

```python
import googlemaps

api_key='AIzaSyDIauObE4LPhLdDpX4KyQK_wcVfo3o741A'
maps = googlemaps.Client(key=api_key)
```

```python
def geo_code(data,columns):
    Lat = [] #Latitude
    Lng = [] #Longitude
    place = []

    for i in data[columns]:
        place.append(i)
        results = maps.geocode(i)
        for result in results:
            Lat.append(result['geometry']['location']['lat'])
            Lng.append(result['geometry']['location']['lng'])
    
    lat_lng = pd.DataFrame([place,Lat,Lng]).T
    lat_lng.columns = ['주소(도로명)','Latitude','Longitude']
    lat_lng = lat_lng[lat_lng['주소(도로명)'].isnull() != True]
    return lat_lng
```

```python
lat_lng = geo_code(market,'주소(도로명)')
lat_lng.head(3)
```

### 전통시장 데이터 합본 (market, totalmarket)

```python
data = pd.merge(left = market, right=lat_lng, on='주소(도로명)')
print(len(data))
data.head(3)
```

```python
totalmarket = pd.read_csv('totalmarket.csv')
print(len(totalmarket))
totalmarket.head(3)
```

```python
data_mr1 = pd.merge(left = data, right=totalmarket, left_on='주소(지번)',right_on='주소')
data_mr2 = pd.merge(left = data, right=totalmarket, left_on='주소(도로명)',right_on='주소')
data_mr3 = pd.merge(left = data, right=totalmarket, left_on='시장(상점가명)',right_on='명칭')
```

```python
data_mr = pd.concat([data_mr1,data_mr2],axis=0, ignore_index=True)
data_1 = pd.concat([data_mr,data_mr3],axis=0,ignore_index=True)

print(len(data_mr))
print(len(data_1))
data_1.head(3)
```

```python
data_1 = data_1.sort_values('시장(상점가명)')
data_1 = data_1.drop_duplicates(['시장(상점가명)'],keep='last')
print(len(data_1))
data_1.head(3)
```
