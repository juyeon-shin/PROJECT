```python
def geo_code(data,columns):
    Lat = [] #Latitude
    Lng = [] #Longitude
    place = []

    for i in data[columns]:
        place.append(i)
        results = maps.geocode(i)
        for result in results:
            Lat.append(result['geometry']['location']['lat'])
            Lng.append(result['geometry']['location']['lng'])
    
    lat_lng = pd.DataFrame([place,Lat,Lng]).T
    lat_lng.columns = ['주소(도로명)','Latitude','Longitude']
    lat_lng = lat_lng[lat_lng['주소(도로명)'].isnull() != True]
    return lat_lng
```

```python
def geo_code(data,columns):
    Lat = [] #Latitude
    Lng = [] #Longitude
    place = []

    for i in data[columns]:
        place.append(i)
        results = maps.geocode(i)
        for result in results:
            Lat.append(result['geometry']['location']['lat'])
            Lng.append(result['geometry']['location']['lng'])
    
    lat_lng = pd.DataFrame([place,Lat,Lng]).T
    lat_lng.columns = ['주소(도로명)','Latitude','Longitude']
    #lat_lng = lat_lng[lat_lng['주소(도로명)'].isnull() != True]
    return lat_lng
```

```python
def mk_merge(data, data_columns, lat_lng):
    mk_data = pd.merge(left = data, right=lat_lng, left_on=data_columns,right_on='주소(도로명)')
    mk_data = mk_data.sort_values('주소(도로명)')
    mk_data = mk_data.drop_duplicates(['주소(도로명)'],keep='last')
    return mk_data
```
